# feature engineering template 

```python
import warnings
warnings.filterwarnings('ignore')
import featuretools as ft
import pandas as pd
import numpy as np
```

```python
df_target = pd.read_csv('dataset/train_target.csv')
df_online_train = pd.read_csv('dataset/online_train.csv')
df_offline_train = pd.read_csv('dataset/offline_train.csv')
```

```python
df_target = df_target[['User_id','Merchant_id','Coupon_id','Date_received','Date','datediff','target']]
df_target['Date_received'] = pd.to_datetime(df_target['Date_received'])
df_target['Date'] = pd.to_datetime(df_target['Date'])
```

```python
df_target['month_received'] = df_target['Date_received'].apply(lambda x : x.month)
```

```python
df_target_5 = df_target[df_target['month_received'] == 5]
```

```python
user = df_target_5[['User_id']].drop_duplicates()
merchant = df_target_5[['Merchant_id']].drop_duplicates()
coupon = df_target_5[['Coupon_id']].drop_duplicates()
```

```python
es = ft.EntitySet(id='offline_test')
```

```python
es = es.entity_from_dataframe(entity_id='user',
                             dataframe=user,
                             index='User_id',
                             variable_types={'User_id':ft.variable_types.Id})

es = es.entity_from_dataframe(entity_id='merchant',
                             dataframe=merchant,
                             index='Merchant_id',
                             variable_types={'Merchant_id':ft.variable_types.Id})

es = es.entity_from_dataframe(entity_id='coupon',
                             dataframe=coupon,
                             index='Coupon_id',
                             variable_types={'Coupon_id':ft.variable_types.Id})

```

```python
df_offline_train['with_coupon'] = df_offline_train['Coupon_id'].notna()
```

```python
df_offline_train['Date_received'] = pd.to_datetime(df_offline_train['Date_received'])
df_offline_train['Date'] = pd.to_datetime(df_offline_train['Date'])
df_offline_train['datediff'] = (df_offline_train['Date'] - df_offline_train['Date_received'] ).apply(lambda x:x.days)
```
